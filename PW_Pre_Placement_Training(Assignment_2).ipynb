```python
#Q1
def generate_pairs(arr):
    pairs = []
    n = len(arr)

    for i in range(n):
        for j in range(i+1, n):
            pairs.append((arr[i], arr[j]))

    return pairs

array = [1,4,3,2]
result = generate_pairs(array)
print(result)
k = len(result)
max=0
res=0
min1=0
min2=0
for i in range(0,k-1,2):
  min1=min(result[i])
  min2=min(result[i+1])
  res = min1+min2
  if(res>max):
    max = res

print(f"The maximum possible sum is {max}")


```

```python
#Q2
def max_unique_candies(candyType):
    unique_candies = len(set(candyType))
    max_candies = len(candyType) // 2
    return min(unique_candies, max_candies)
candyType = [1, 1, 2, 2, 3, 3, 4, 6, 6, 7]
result = max_unique_candies(candyType)
print(result)
```

```python
#Q3
def findLHS(nums):
    freq_dict = {}
    max_length = 0

    # Count the occurrences of each number
    for num in nums:
        freq_dict[num] = freq_dict.get(num, 0) + 1

    # Find the length of the longest harmonious subsequence
    for num in freq_dict:
        if num + 1 in freq_dict:
            subseq_length = freq_dict[num] + freq_dict[num + 1]
            if subseq_length > max_length:
                max_length = subseq_length

    return max_length
nums = [1, 3, 2, 2, 5, 2, 3, 7]
result = findLHS(nums)
print(result)
```

```python
#Q5
data = [2, 4, 3, 4, 9, 1, 7, 8, 5, 4]
m = len(data)
maximum = 0
a, b, c = 0, 0, 0

for i in range(0,m-2):
    for j in range(i + 1, m-1 ):
        for k in range(j+1,m):
            p = data[i] * data[j] * data[k]
            if p > maximum:
                maximum = p
                a = i
                b = j
                c = k

print(f"The product of {data[a]} X {data[b]} X {data[c]} is = {maximum}")
```

```python
#Q6
def search(nums, target):
    left = 0
    right = len(nums) - 1

    while left <= right:
        mid = left + (right - left) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1
nums = [-1,0,3,5,9,12]
target = 9
result = search(nums, target)
print(result)
```

```python
#Q7
def isMonotonic(nums):
    increasing = True
    decreasing = True

    for i in range(1, len(nums)):
        if nums[i] < nums[i - 1]:
            increasing = False
        if nums[i] > nums[i - 1]:
            decreasing = False

    return increasing or decreasing
nums = [1, 2, 3, 4, 5]
result = isMonotonic(nums)
print(result)
```
